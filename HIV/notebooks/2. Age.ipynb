```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
hiv_df = pd.read_csv("../data/hiv.csv")
hiv_df.head()
```

```python
age_df = hiv_df.loc[hiv_df['Demographic Category'] == "Age Group (years)"]
age_df
```

```python
plt.bar(age_df['Demographic Type'], age_df['HIV/AIDS Population Rate'])
plt.show()
```

```python
traces = ['HIV (non-AIDS) Population Rate', 'AIDS Population Rate']
x_cats = age_df['Demographic Type']
x_values = np.arange(len(x_cats))

for i in range(len(traces)):
    plt.bar(x_values+.2*(i-.5), age_df[traces[i]], width=.2)

plt.xticks(x_values, x_cats, rotation=30)
plt.show()
```

```python
age_df = age_df.dropna().drop(age_df.iloc[:, [0, 2, 3, 5, 6, 8, 9]], axis=1)
age_df.to_csv("../data/cleaned/age.csv", index=False)
```
