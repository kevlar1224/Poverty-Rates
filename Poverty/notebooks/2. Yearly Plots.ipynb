```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
year_df = pd.read_csv("../data/byYear/total.csv")
year_df.head()
```

```python
df = year_df

plt.plot(df['Year'], df['Percent'])
plt.show()
```

```python
race_df = pd.read_csv("../data/byYear/race.csv")
race_df.head()
```

```python
df = race_df
traces = df.groupby('Race').sum().index

for trace in traces:
    selected_trace = df.loc[df['Race'] == trace]
    plt.plot(selected_trace['Year'], selected_trace['Percent'])
    plt.show()
```

```python
type_df = pd.read_csv("../data/byYear/type.csv")
type_df.head()
```

```python
df = type_df
traces = df.groupby('Type').sum().index

for trace in traces:
    selected_trace = df.loc[df['Type'] == trace]
    plt.plot(selected_trace['Year'], selected_trace['Percent'])
    plt.show()
```
