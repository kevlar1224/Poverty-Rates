```python
import pandas as pd
pd.set_option("display.max_rows", 5000)
```

```python
base_df = pd.read_csv("../data/infant-mortality.csv")
base_df = base_df.rename(columns = {"Race_Ethnicity":'Race', 'Total_LiveBirths_Denominator': 'Total Live Births'})
base_df.head()
```

```python
race_total = base_df.loc[base_df['Race'] != 'Total']\
            .loc[base_df['Gender'] == 'Total']\
            .loc[base_df['Age at death'] == 'Total']\
            .drop("Gender", axis=1)\
            .drop("Age at death", axis=1)

race_total.head()
```

```python
grouped_race = race_total.groupby('Race', as_index=False).sum()
grouped_race['Rate'] = round(grouped_race['Count']*1000/grouped_race['Total Live Births'], 2)
grouped_race = grouped_race.drop("Year", axis=1)
grouped_race.to_csv("../data/race/total.csv", index=False)
```

```python
year_totals = base_df.loc[base_df['Race'] == 'Total']\
            .loc[base_df['Gender'] == 'Total']\
            .loc[base_df['Age at death'] == 'Total']\
            .drop("Race", axis=1)\
            .drop("Gender", axis=1)\
            .drop("Age at death", axis=1)

year_totals.to_csv("../data/byYear/total.csv", index=False)
year_totals.head()
```

```python
new_races = ['American Indian and/or Alaska Native', 'Asian', 'Black or African American', 
             'Hispanic', 'Mixed Race', 'Native Hawaiian and Other Pacific Islander', 
             'White', 'Not Reported', 'Other']
```

```python
race_year = base_df.loc[base_df['Gender'] == 'Total']\
            .loc[base_df['Age at death'] == 'Total']\
            .loc[base_df['Race'] != 'Total']\
            .drop('Gender', axis=1)\
            .drop('Age at death', axis=1)

race_year.to_csv("../data/byYear/race.csv", index=False)
race_year.head()
```

```python
gender_year = base_df.loc[base_df['Gender'] != 'Total']\
            .loc[base_df['Race'] == 'Total']\
            .loc[base_df['Age at death'] == 'Total']\
            .drop('Race', axis=1)\
            .drop('Age at death', axis=1)

gender_year.to_csv("../data/byYear/gender.csv", index=False)
gender_year.head()
```

```python
age_year = base_df.loc[base_df['Age at death'] != 'Total']\
            .loc[base_df['Race'] == 'Total']\
            .loc[base_df['Gender'] == 'Total']\
            .drop('Race', axis=1)\
            .drop('Gender', axis=1)

age_year.to_csv("../data/byYear/age.csv", index=False)
age_year.head()
```

```python
race_gender = base_df.loc[base_df['Race'] != 'Total']\
            .loc[base_df['Gender'] != 'Total']\
            .loc[base_df['Age at death'] == 'Total']\
            .drop('Age at death', axis=1)

race_gender = race_gender.groupby(['Race', 'Gender'], as_index=False).sum().drop('Year', axis=1)
race_gender['Rate'] = round(race_gender['Count']*1000/race_gender['Total Live Births'], 2)
race_gender.to_csv("../data/race/gender.csv", index=False)
race_gender.head()
```

```python
race_age = base_df.loc[base_df['Race'] != 'Total']\
            .loc[base_df['Age at death'] != 'Total']\
            .loc[base_df['Gender'] == 'Total']\
            .drop('Gender', axis=1)

race_age = race_age.groupby(['Race', 'Age at death'], as_index=False).sum().drop('Year', axis=1)
race_age['Rate'] = round(race_age['Count']*1000/race_age['Total Live Births'], 2)
race_age.to_csv("../data/race/age.csv", index=False)
race_age.head()
```

```python
gender_age = base_df.loc[base_df['Gender'] != 'Total']\
            .loc[base_df['Age at death'] != 'Total']\
            .loc[base_df['Race'] == 'Total']\
            .drop('Race', axis=1)

gender_age = gender_age.groupby(['Gender', 'Age at death'], as_index=False).sum().drop('Year', axis=1)
gender_age['Rate'] = round(gender_age['Count']*1000/gender_age['Total Live Births'], 2)
gender_age.to_csv("../data/gender/age.csv", index=False)
gender_age.head()
```
