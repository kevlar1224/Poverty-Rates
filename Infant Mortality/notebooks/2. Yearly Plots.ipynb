```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
year_totals = pd.read_csv("../data/byYear/total.csv")
year_totals
```

```python
plt.plot(year_totals['Year'], year_totals['Count'])
plt.show()
```

```python
plt.plot(year_totals['Year'], year_totals['Rate'])
plt.show()
```

```python
race_year = pd.read_csv("../data/byYear/race.csv")
race_year.head()
```

```python
race_list = race_year.groupby("Race").sum()['Count'].keys().tolist()

for i in race_list:
    selected_race = race_year.loc[race_year['Race'] == i]
    plt.plot(selected_race['Year'], selected_race['Count'])
    plt.title(i)
    plt.show()
```

```python
gender_year = pd.read_csv("../data/byYear/gender.csv")
gender_year.head()
```

```python
selected_gender = gender_year.loc[gender_year['Gender'] == "Female"]
plt.plot(selected_gender['Year'], selected_gender['Count'])
plt.title("Female")
plt.show()
```

```python
selected_gender = gender_year.loc[gender_year['Gender'] == "Male"]
plt.plot(selected_gender['Year'], selected_gender['Count'])
plt.title("Male")
plt.show()
```

```python
age_year = pd.read_csv("../data/byYear/age.csv")
age_year.head()
```

```python
selected_age = age_year.loc[age_year['Age at death'] == "Neonatal"]
plt.plot(selected_age['Year'], selected_age['Count'])
plt.title("Neonatal")
plt.show()
```

```python
selected_age = age_year.loc[age_year['Age at death'] == "PostNeonatal"]
plt.plot(selected_age['Year'], selected_age['Count'])
plt.title("PostNeonatal")
plt.show()
```
